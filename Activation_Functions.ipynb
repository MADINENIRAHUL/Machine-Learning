### Here every activation Function is done in single code snippet


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special

# Activation Functions
def linear(x):
    return x

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

def prelu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * x)

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Normalize for stability
    return exp_x / np.sum(exp_x)

def gelu(x):
    return x * 0.5 * (1 + scipy.special.erf(x / np.sqrt(2)))

def mish(x):
    return x * np.tanh(np.log1p(np.exp(x)))  # log1p(x) = log(1 + x) for stability

def softplus(x):
    return np.log1p(np.exp(x))

def bent_identity(x):
    return ((np.sqrt(x**2 + 1) - 1) / 2) + x

def hard_sigmoid(x):
    return np.maximum(0, np.minimum(1, 0.2 * x + 0.5))

def hard_swish(x):
    return x * np.maximum(0, np.minimum(1, (x + 3) / 6))

def maxout(x):
    w1, b1 = 1, 0  # Example parameters
    w2, b2 = -1, 1
    return np.maximum(w1 * x + b1, w2 * x + b2)

# Generate x values
x = np.linspace(-5, 5, 400)

# Define function names and their corresponding functions
functions = {
    "Linear": linear,
    "Sigmoid": sigmoid,
    "Tanh": tanh,
    "ReLU": relu,
    "Leaky ReLU": leaky_relu,
    "PReLU": prelu,
    "ELU": elu,
    "Softmax": lambda x: softmax(x),
    "GELU": gelu,
    "Mish": mish,
    "Softplus": softplus,
    "Bent Identity": bent_identity,
    "Hard Sigmoid": hard_sigmoid,
    "Hard Swish": hard_swish,
    "Maxout": maxout
}

# Plot each function
plt.figure(figsize=(14, 16))

for i, (name, func) in enumerate(functions.items(), 1):
    plt.subplot(5, 3, i)
    plt.plot(x, func(x), label=name, color='b')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.title(name)
    plt.legend()
    plt.grid()

plt.tight_layout()
plt.show()
```
